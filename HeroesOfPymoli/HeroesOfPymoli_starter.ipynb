```python
#Dependencies and setup
import pandas as pd

#load file
file = "../Resources/purchase_data.csv"

#read in CSV file in Pandas data frame
Purchase_df = pd.read_csv(file)
```

```python
#get total number of unique values of players
Total_Players = len(Purchase_df["SN"].value_counts())

#Create data frame with total count of players 
Count = pd.DataFrame({"Total Players":[Total_Players]})
Count
```

```python
#Calculate Items, Average, Number of Purchases and Total Revenue
Items = len(Purchase_df["Item ID"].value_counts())
Average_Price = (Purchase_df["Price"]).mean()
Number_Of_Purchases = (Purchase_df["Purchase ID"].count())
Total_Revenue = (Purchase_df["Price"]).sum()

#Create Data Frame to store the values
Summary_df = pd.DataFrame({"Number of Unique Items": [Items],
                            "Average Price": [Average_Price],
                            "Number Of Purchases": [Number_Of_Purchases],
                            "Total Revenue": [Total_Revenue]})

#Adjust the formating style of the output
Summary_df.style.format({'Average Price':"${:,.2f}",
                         'Total Revenue': '${:,.2f}'})
```

```python
#Group purchase_data by Gender
Group_Gender = Purchase_df.groupby('Gender')

#Count the total of screen names in column "SN" by gender
Count_Gender = Group_Gender.nunique()["SN"]

#Calculate Percentage for Genders
Percentage = Count_Gender / Total_Players * 100

#Create Data Frame to output gender Demographics
Gender_Demographics = pd.DataFrame({"Total Count": Count_Gender,
                                 "Percentage of Players": Percentage})

#Get rid of index
Gender_Demographics.index.name = None

#Gender_Demographics.sort_values(["Total Count"], ascending = False)
Gender_Demographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}%"})
```

```python
#Count of purchases by gender
Purchase_Count = Group_Gender['Purchase ID'].count()

#Average by gender
Avg_Purchase = Group_Gender['Price'].mean()

#Average total by gender
Avg_Purchase_Total = Group_Gender['Price'].sum()

#Average per person
Avg_Per_Person = Avg_Purchase_Total/Count_Gender

# Create data frame with obtained values 
Purchasing_df = pd.DataFrame({"Purchase Count": Purchase_Count, 
                                    "Average Purchase Price": Avg_Purchase,
                                    "Total Purchase Value":Avg_Purchase_Total,
                                    "Avg Purchase Total per Person": Avg_Per_Person})

# Provide index in top left as "Gender"
Purchasing_df.index.name = "Gender"

# Format with currency style
Purchasing_df.style.format({"Total Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

```python
# Establish bins for ages
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99999]
age_group = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Segment and sort age values into bins established above
Purchase_df["Age Group"] = pd.cut(Purchase_df["Age"],age_bins, labels=age_group, include_lowest=True)
```

```python
# Create new data frame with the added "Age Group" and group it
age_grouped = Purchase_df.groupby("Age Group")

# Count total players by age category
total_count_age = age_grouped["SN"].nunique()

# Calculate percentages by age category 
percentage_by_age = (total_count_age / Total_Players) * 100

# Create data frame with obtained values
age_demographics = pd.DataFrame({"Total Count": total_count_age, "Percentage of Players": percentage_by_age})

# Format the data frame with no index 
age_demographics.index.name = None

# Format percentage with two decimal places 
age_demographics.style.format({"Percentage of Players":"{:,.2f}%"})
```
